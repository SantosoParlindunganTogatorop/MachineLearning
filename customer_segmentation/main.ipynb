```python
X1 = df.loc[:,['Age','Spending Score (1-100)']].values
```

```python
X1 = df.loc[:,['Age','Spending Score (1-100)']].values

from sklearn.cluster import KMeans
kmean_inirtia = []

for i in range(1,11):
    kmean = KMeans(n_clusters=i, init='k-means++')
    kmean.fit(X1)
    kmean_inirtia.append(kmean.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(kmean_inirtia, linewidth=2, color='red', marker='8')
plt.xlabel('K Values')
plt.ylabel('Inirtia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(X1)

print(kmeans.cluster_centers_)
```

```python
plt.scatter(X1[:,0], X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('Clsuters of Customer')  
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
X2 = df.loc[:,['Annual Income (k$)','Spending Score (1-100)']].values

from sklearn.cluster import KMeans
kmean_inirtia = []

for i in range(1,11):
    kmean = KMeans(n_clusters=i, init='k-means++')
    kmean.fit(X2)
    kmean_inirtia.append(kmean.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(kmean_inirtia, linewidth=2, color='red', marker='8')
plt.xlabel('K Values')
plt.ylabel('Inirtia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(X2)

print(kmeans.cluster_centers_)
```

```python
plt.scatter(X2[:,0], X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('Clsuters of Customer')
plt.xlabel('Annual Income ($)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
X3 = df.loc[:,['Age','Annual Income (k$)','Spending Score (1-100)']].values
kmean_inirtia = []

for i in range(1,11):
    kmean = KMeans(n_clusters=i, init='k-means++')
    kmean.fit(X3)
    kmean_inirtia.append(kmean.inertia_)

plt.figure(figsize=(15,6))
plt.plot(range(1,11), kmean_inirtia, marker='o', color='red', linewidth=2)
plt.title('Elbowh Method')
plt.xlabel('n Cluster')
plt.ylabel('Inertia')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

kmean = KMeans(n_clusters=4)
ypred = kmean.fit_predict(X3)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X3[:,0], X3[:,1], X3[:,2], c=ypred, s=50, cmap='viridis', alpha=0.7)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
plt.title('Customer Segmentation 3D')
plt.show()
```
